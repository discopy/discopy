```python
x, y = Ty('x'), Ty('y')
f, g = Box('f', x, x @ y), Box('g', y @ x, x)
d = f @ Id(x) >> Id(x) @ g
diagram = (d @ d.dagger() >> d.dagger() @ d)
diagram.draw()
diagram.normal_form().draw()
pass
```

```python
spiral_steps = list(moncat._spiral(5).normalize())
@interact
def normalise(step=(0, len(spiral_steps) - 1, 1)):
    spiral_steps[step].draw()
```

```python
a, b, c, d = Ty('a'), Ty('b'), Ty('c'), Ty('d')
snake = Box('f', a @ b, c @ d)\
    .transpose_l().transpose_l()\
    .transpose_r().transpose_r()\
    .transpose_l().transpose_r()
snake_steps = list(snake.normalize())
@interact
def normalise(step=(0, len(snake_steps) - 1, 1)):
    snake_steps[step].draw()
```
